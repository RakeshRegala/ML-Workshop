#5. Create a DataFrame for employees:
data = {
 'Department': ['HR', 'HR', 'IT', 'IT', 'Finance'],
 'Employee': ['Alice', 'Bob', 'Carol', 'David', 'Eve'],
 'Salary': [50000, 45000, 60000, 65000, 70000]
}
â€¢ Group the data by Department and compute the total and mean salary for each
department.

```python
data = { 'Department': ['HR', 'HR', 'IT', 'IT', 'Finance'], 'Employee': ['Alice', 'Bob', 'Carol', 'David', 'Eve'], 'Salary': [50000, 45000, 60000, 65000, 70000] }
df = pd.DataFrame(data)
grouped = df.groupby('Department')
total_salary = grouped['Salary'].sum()
mean_salary = grouped['Salary'].mean()
print(total_salary)
print(mean_salary)
```

#7. Given a small dataset, clean and display basic statistics using Pandas.


```python

# Task 7
# Given a small dataset, clean and display basic statistics
data = {
    'Name': ['Alice', 'Bob', 'Carol', 'David', 'Eve'],
    'Age': [25, None, 30, 35, None],
    'Salary': [50000, 45000, 60000, 65000, 70000]
}
small_df = pd.DataFrame(data)

# Fill missing values in Age with the mean age
mean_age = small_df['Age'].mean()
small_df['Age'] = small_df['Age'].fillna(mean_age)

# Display basic statistics
statistics = small_df.describe()

# Display results
print("Cleaned Dataset:")
print(small_df)
print("\nBasic Statistics of Small Dataset:")
print(statistics)
```

#9. Data Transformation
â€¢ Load a dataset containing daily temperatures and dates. Perform the following:
o Convert the date column into a DateTime object.
o Create a new column that categorizes temperatures into High, Medium, or
Low.

```python
data = {
    'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04'],
    'Temperature': [30, 22, 15, 10]
}
temp_df = pd.DataFrame(data)

# Convert the date column into a DateTime object
temp_df['Date'] = pd.to_datetime(temp_df['Date'])

# Create a new column that categorizes temperatures into High, Medium, or Low
def categorize_temperature(temp):
    if temp > 25:
        return 'High'
    elif 15 <= temp <= 25:
        return 'Medium'
    else:
        return 'Low'

temp_df['Temperature_Category'] = temp_df['Temperature'].apply(categorize_temperature)
temp_df
```
