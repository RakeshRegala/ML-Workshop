#Top 10 Youtubers

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = '/content/youtube_subscribers_data.csv'
youtube_data = pd.read_csv(file_path)

# Get the top 10 channels by subscriber count
top_10_channels = youtube_data.nlargest(10, 'Subscribers (millions)')
# Visualization: Top 10 Channels by Subscriber Count
plt.figure(figsize=(10, 6))
sns.barplot(
    data=top_10_channels,
    y="Name",
    x="Subscribers (millions)",
    palette="viridis"
)
plt.title("Top 10 YouTube Channels by Subscriber Count", fontsize=14)
plt.xlabel("Subscribers (millions)", fontsize=12)
plt.ylabel("Channel Name", fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

#Distribution of Channels by Subscriber Count Range

```python
# Categorize subscriber counts into ranges
bins = [0, 50, 100, 150, 200, 250, 300, 350]
labels = ['<50M', '50-100M', '100-150M', '150-200M', '200-250M', '250-300M', '300M+']
youtube_data['Subscriber Range'] = pd.cut(youtube_data['Subscribers (millions)'], bins=bins, labels=labels)

# Count channels in each range
range_counts = youtube_data['Subscriber Range'].value_counts().sort_index()

# Visualization: Channel Count by Subscriber Range
plt.figure(figsize=(10, 6))
sns.barplot(x=range_counts.index, y=range_counts.values, palette='Blues')
plt.title('Channel Count by Subscriber Range', fontsize=14)
plt.xlabel('Subscriber Range', fontsize=12)
plt.ylabel('Channel Count', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
